```python
# Import library pandas sebagai 'pd'
import pandas as pd

# Keterangan
# 'pandas' merupakan library python yang digunakan untuk analisis data, manipulasi data, dan pengolahan data.
```

```python
# Disini saya menggunakan dataset 'Letter Recognition' dalam format csv.
# File tersebut kemudian saya upload di Github (sadamapeach) untuk mendapatkan URL data.
# Terdapat 17 atribut dalam dataset yang saya gunakan.

# Link dataset
url = "https://raw.githubusercontent.com/sadamapeach/Machine-Learning/main/Letter_Recognition.csv"

# Atribut dataset
names = ['letter', 'xbox', 'ybox', 'width', 'height', 'onpix', 'xbar', 'ybar', 'x2bar', 'y2bar', 'xybar', 'x2ybar', 'xy2bar', 'xedge', 'xedgey', 'yedge', 'yedgex']

# pd.read_csv = library pandas untuk membaca dataset dari berkas csv
dataset = pd.read_csv(url, names=names)
```

```python
# Import library matplotlib
import matplotlib.pyplot as plt

# Keterangan
# 'matplotlib.pyplot' merupakan library python yang digunakan untuk membuat grafik dan plot dalam berbagai format.
```

```python
# Dataset Boxplot
dataset.plot(kind='box', subplots=True, layout=(20,4), sharex=False, sharey=False, figsize=(10, 50))
plt.show()

# Keterangan
# kind='box' untuk menggambar box-plot
# subplots=True menghasilkan subplots yang terpisah untuk setiap kolom dataset
# layout(20,4) mengatur tata letak subplots menjadi 20 baris dan 4 kolom
# sharex & sharey memastikan sumbu x dan y tidak dibagi antar subplots
# figsize untuk mengatur ukuran gambar
```

Keterangan:
*   Boxplot memvisualisasikan data numerik berdasarkan nilai min, max, Q1, Q2, Q3, beserta outliernya.
*   Setiap data memiliki nilai min = 0 dan max = 15.
*   Untuk nilai Q1, Q2, dan Q3 dapat dilihat pada ringkasan statistik (No.4a).

Visualisasi @atribut:
*   Atribut 'xbox' memiliki 7 outlier atas.
*   Atribut 'ybox' tidak memiliki outlier.
*   Atribut 'width' memiliki 1 outlier bawah dan 6 outlier atas.
*   Atribut 'height' memiliki 4 outlier atas.
*   Atribut 'onpix' memiliki 6 outlier atas.
*   Atribut 'xbar' memiliki 3 outlier bawah dan 4 outlier atas.
*   Atribut 'ybar' memiliki 2 outlier bawah dan 2 outlier atas.
*   Atribut 'x2bar' memiliki 5 outlier atas.
*   Atribut 'y2bar' memiliki 4 outlier atas.
*   Atribut 'xybar' memiliki 3 outlier bawah dan 1 outlier atas.
*   Atribut 'x2ybar' memiliki 1 outlier bawah dan 3 outlier atas.
*   Atribut 'xy2bar' memiliki 4 outlier bawah dan 3 outlier atas.
*   Atribut 'xedge' memiliki 7 outlier atas.
*   Atribut 'xedgey' memiliki 7 outlier bawah dan 5 outlier atas.
*   Atribut 'yedge' memiliki 6 outlier atas.
*   Atribut 'yedgex' memiliki 4 outlier bawah dan 3 outlier atas.
