```python
# Nama    : Oktaviana Sadama Nur Azizah
# NIM     : 24060121130060
# Dataset : Iris
# Link    : http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data
```

```python
# Import library pandas sebagai 'pd'
import pandas as pd

# Keterangan
# 'pandas' merupakan library python yang digunakan untuk analisis data, manipulasi data, dan pengolahan data.
```

```python
# Link dataset iris
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Atribut dataset
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']

# pd.read_csv = library pandas untuk membaca dataset dari berkas csv
dataset = pd.read_csv(url, names=names)
```

```python
dataset.head(5)
```

```python
dataset.shape
```

```python
# Train-Test Split
from sklearn import model_selection
```

```python
# NO.1 MEMBUAT VALIDASI DATASET
# 'dataset.values' disimpan dalam variabel array
array = dataset.values

# Membagi data dalam variabel x dan y
# Variabel X berisi data fitur index 0 sampai 3 yang digunakan sebagai inputan
X = array[:,0:4]
# Variabel Y berisi data index ke 4
Y = array[:,4]

# Melatih model dengan 20% data keseluruhan
validation_size = 0.20

# Mengacak data untuk memastikan bahwa hasil pembagian data konsisten
seed = 7

# 'train_test_split' merupakan pustaka model_selection
# Fungsi: membagi data pelatihan dan validasi berdasarkan parameter yang diatur
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

```python
print("X_train :", X_train.shape)
print("X_validation :", X_validation.shape)
print("Y_train :", Y_train.shape)
print("Y_validation :", Y_validation.shape)

# Keterangan
# 80% data training dan 20% data validasi
```

```python
# NO.2 K-FOLDS CROSS VALIDATION
# Melipat data dan mengiterasi eksperimen sebanyak K
seed = 7

# Matriks evaluasi yang digunakan adalah 'accuracy'
scoring = 'accuracy'
```

```python
# NO.3 MEMBANGUN MODEL
# Mendefinisikan model-model yang akan digunakan

# Import library berisi algoritma klasifikasi yang akan digunakan
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.tree import DecisionTreeClassifier

models = []

# Array assosiatif algoritma klasifikasi
models.append(('Discriminant', QuadraticDiscriminantAnalysis()))
models.append(('Gaussian Process', GaussianProcessClassifier()))
models.append(('Tree', DecisionTreeClassifier()))
```

```python
# Evaluate each model in turn
# Menyimpan hasil evaluasi dari @model
results = []

# Menyimpan nama-nama model
names = []

# Validasi menggunakan K-FOLD VALIDATION
for name, model in models:

  # Membagi data menjadi training dan validation
  # k = 10 dan random seed = 7
  kfold = model_selection.KFold(n_splits=10, random_state=seed, shuffle=True)

  # Mengukur kinerja model dengan data latih
  # Menghitung akurasi dari tiap iterasi
  cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)

  # Menambahkan nilai iterasi tiap akurasi ke variabel 'results'
  results.append(cv_results)

  # Menyimpan nama model
  names.append(name)

  # Print pesan (nama model, rata-rata, standar deviasi)
  msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
  print(msg)
```

```python
# NO.4 MEMILIH MODEL TERBAIK DISCRIMINANT
# Import library
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Model terbaik Discriminant
discriminant = QuadraticDiscriminantAnalysis()

# Melatih model discriminant dengan data latih X_train dan Y_train
discriminant.fit(X_train, Y_train)

# Memprediksi model discriminant yang sudah dilatih
predictions = discriminant .predict(X_validation)
print(accuracy_score(Y_validation, predictions))

print(confusion_matrix(Y_validation, predictions))
print(classification_report(Y_validation, predictions))
```

```python
# NO.4 MEMILIH MODEL TERBAIK GAUSSIAN
# Import library
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Model terbaik Gaussian
gaussian = GaussianProcessClassifier()

# Melatih model Gaussian dengan data latih X_train dan Y_train
gaussian.fit(X_train, Y_train)

# Memprediksi model Gaussian yang sudah dilatih
predictions = gaussian.predict(X_validation)
print(accuracy_score(Y_validation, predictions))

print(confusion_matrix(Y_validation, predictions))
print(classification_report(Y_validation, predictions))
```

```python
# Meskipun algoritma klasifikasi discriminant dan gaussian process
# memiliki nilai akurasi yang sama, yaitu 0.983333 (0.033333)
# tetapi setelah dilakukan pengujian model terbaik,
# didapati bahwa model Discriminant memiliki nilai yang lebih tinggi
# dibandingkan dengan model Gaussian.
```
